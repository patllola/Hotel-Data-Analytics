### Data Tranformation

There are various data transformation that you may have to perform based on the need. Few of the data transformation are given bvelow:

1. Creating new column ( Creating new column and storing the new values)
2. Normalizaton  ( Converting the data into another form)
3. Merge the Data (Merging two table)
4. Aggregation (Mean, std, sum, max, min )

```python
df_agg_bookings
```

```python
# occupany percentage
25/30
```

```python
df_agg_bookings.head()
```

```python
df_agg_bookings["occ_per"]=df_agg_bookings["successful_bookings"]/df_agg_bookings["capacity"]
```

```python
df_agg_bookings
```

```python
#  Tranfering the data in understandable format and adding the percentage value
def fun(occ_per):
    x=round(occ_per*100,2)
    return x
df_agg_bookings["occ_per"]=df_agg_bookings["occ_per"].apply(fun)
```
